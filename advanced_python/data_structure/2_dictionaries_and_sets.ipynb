we focused on sequence patterns only, but different types of patterns
can be combined and nested. Thanks to destructuring, pattern matching is a power‐
ful tool to process records structured like nested mappings and sequences, which we

```python
def get_creators(record: dict) -> list:
    match record:
        case {'type': 'book', 'api': 2, 'authors': [*names]}:
            return names
        case {'type': 'book', 'api': 1, 'author': name}:
            return [name]
        case {'type': 'book'}:
            raise ValueError(f"Invalid 'book' record: {record!r}")
        case {'type': 'movie', **details}:
            return [details.items()]
        case _:
            raise ValueError(f'Invalid record: {record!r}')

get_creators({'type': 'book', 'api': 1, 'author': 'mike'})
```

If you override __eq__ to compare internal state (like name), you must override __hash__ to use the same state — and that state must not change during the object’s lifetime.

```python
class UltimatePerson:
    def __init__(self, name):
        self.name = name

    def __eq__(self, other):
        return isinstance(other, UltimatePerson) and self.name == other.name

    def __hash__(self):
        return hash(self.name)

name_1 = ('ahmed')
name_2 = ("ahmed")
name_1 == name_2
```

defaultdict is good when you want Python to automatically create missing entries.

when instantiating a defaultdict, you provide a callable to
produce a default value whenever __getitem__ is passed a nonexistent key argument.
For example, given a defaultdict created as dd = defaultdict(list), if 'new-key'
is not in dd, the expression dd['new-key'] does the following steps:
1. Calls list() to create a new list.
2. Inserts the list into dd using 'new-key' as key.
3. Returns a reference to that list

The mechanism that makes defaultdict work by calling default_factory is the
__missing__ special method,

Underlying the way mappings deal with missing keys is the aptly named __missing__
method. This method is not defined in the base dict class, but dict is aware of it: if
you subclass dict and provide a __missing__ method, the standard dict.__geti
tem__ will call it whenever a key is not found, instead of raising KeyError.

```python
class StrDict(dict):
    def __missing__(self,key):
        if isinstance(key,str):
            raise KeyError(key)
        return self[str(key)]

    def get(self,key,default=None):
        try:
            return self[key]
        except KeyError:
            return default
    def __contains__(self, key): #for in operator
        return key in self.keys() or str(key) in self.keys()

```

Why?

Because Python searches for variables in layers (scopes):

Local scope (inside the function)

Global scope

Built-ins (Python's predefined names like len, print, etc.)

This rule is called LEGB:

```python
from collections import ChainMap
import builtins

x = 10

def test():
    x = 5
    pylookup = ChainMap(locals(), globals(), vars(builtins))
    
    print(pylookup['x'])     # local
    print(pylookup['len'])   # builtin

test()
```

```python
def test():
    x = 5
    pylookup = ChainMap(locals(), globals(), vars(builtins))

    print("local x:", pylookup.maps[0]['x'])
    print("global x:", pylookup.maps[1]['x'])

test()
```
