# Python special methods, also known as dunder (double-underscore) or magic methods, are built-in methods with reserved names that allow you to customize how your objects behave with standard Python operations, built-in functions, and operators.


```python
name = "Alice"

print(f"{name}")    # Alice
print(f"{name!r}")  # 'Alice'

#this gets the __reper__
```

```python
class FruitsBox:
    def __init__(self):
        self.fruits = ['banana','apple']

    def __getitem__(self,index):
        return self.fruits[index]

a = FruitsBox()
print(a[1])
```

```python
class FruitsBox:
    def __init__(self):
        self.fruits = ['banana','apple']

    def __getitem__(self,index):
        if isinstance(index,slice):
            return self.fruits[index.start:index.stop:index.step]
        else:
            return self.fruits[index]

a = FruitsBox()
print(a[0:2:1])
```

They are immutable, like regular tuples, and are highly memory efficient, making them ideal for creating simple, lightweight data structures without the overhead of a full class definition

Python can read the length directly from memory (C-level structure) without calling a normal method. That makes it very fast.

So len(x) is optimized at a lower level in CPython.

```python
class FrenchDeck:
    ranks = [str(n) for n in range(2,11)] + list('JQKA')  #JACK,QUEEN,KING,ACE
    suits = 'spades diamonds clubs hearts'.split()

    def __init__(self):
        self._cards = [Card(rank,suit) for suit in self.suits for rank in self.ranks]

    def __len__(self):
        return len(self._cards)

    def __getitem__(self,order:int | slice)-> Card | list[Card]:
        return self._cards[order]

deck = FrenchDeck()
len(deck),deck[0],deck[0:10]
```

our deck also iterable

```python
from itertools import islice

for card in islice(reversed(deck), 5):
    print(card)
```

```python
rev_deck = reversed(deck)
i = 0
while i < 5:
    try:
        card = next(rev_deck)
        print(card)
        i+=1
    except StopIteration:
        print(f'out of bounds, the size of deck: {len(deck)}')
        break
```

```python
suit_value = dict(spades=3,hearts=2,diamonds=1,clubs=0)
def spades_hearts_diamonds_clubs(card):
    #class variables are shared by all instance (ranks and suits)
    rank_value = FrenchDeck.ranks.index(card.rank)
    return rank_value * len(suit_value) + suit_value[card.suit]

spades_hearts_diamonds_clubs(Card(rank='A', suit='hearts'))
```

Python uses len(x) instead of x.len() because len is a built-in function that is part of Python’s data model, not a regular method. Although it looks like a simple function, it actually calls the special method x.__len__() behind the scenes. This design makes built-in types like lists and strings faster (since their length can be read directly from memory), keeps the interface consistent across different types, and reflects Python’s philosophy that “practicality beats purity.” So len is used as an example to show how Python balances performance, consistency, and clean design while still allowing custom objects to participate through special (magic) methods.
