### Functions in Python are first-class objects.
 Programming language researchers define a “first-class object” as a program entity that can be:

• Created at runtime

• Assigned to a variable or element in a data structure

• Passed as an argument to a function

• Returned as the result of a function

```python
#First class nature

def factorial(n):
    return 1 if n <= 1 else n* factorial(n-1)
factorial(5)
fac = factorial
facc = map(factorial,(range(5)))

fac,facc,fac(3),next(facc)
```

### Nine Flavors of Callable Objects in Python:

Callable objects → Anything you can use with (); check with callable(obj).

User-defined functions → Created with def or lambda.

Built-in functions → Implemented in C, e.g., len(), time.strftime().

Built-in methods → Methods implemented in C, e.g., dict.get().

Methods → Functions defined inside a class body.

Classes → Calling a class runs __new__ → __init__ → returns instance.

Class instances → Instances with a __call__ method can be invoked like functions.

Generator functions → Use yield; calling them returns a generator object.

Native coroutine functions → Defined with async def; calling returns a coroutine (Python 3.5+).

Asynchronous generator functions → async def with yield; calling returns an async generator (Python 3.6+).

### User-Defined Callable Types

Not only are Python functions real objects, but arbitrary Python objects may also be
made to behave like functions. Implementing a __call__ instance method is all it
takes.

```python
class TaskManager:
    def __init__(self):
        self._tasks = []          # Protected list of tasks (convention: internal use)
        self.__completed = 0      # Private counter (name mangling prevents accidental access)

    def add_task(self, *task):
        self._tasks.extend(task)
        print(f"Task added: {task}")

    def __call__(self, complete_task=True):
        """
        Calling the instance completes one task if available.
        """
        if self._tasks and complete_task:
            task = self._tasks.pop(0)
            self.__completed += 1
            print(f"Completed task: {task}")
        else:
            print("No tasks to complete.")

    def status(self):
        print(f"Pending tasks: {len(self._tasks)}")
        print(f"Completed tasks: {self.__completed}")

task = TaskManager()
task.add_task('design','architicture')
```

```python
import random
class PickFruit:
    def __init__(self,fruits=None):
        self._fruits = list(fruits) if fruits else []

    def pick(self):
        random.shuffle(self._fruits)
        try:
            return self._fruits.pop()
        except:
            return f"list is empty"
    
    def __call__(self):
        return self.pick()

pf=PickFruit(['banana','apple','orange'])
for item in pf._fruits:
    print(item)
print(pf())
```

```python
def tag(name, *content, class_=None, **attrs):
    """Generate one or more HTML tags"""
    if class_ is not None:
        attrs['class'] = class_
        
    attr_pairs = (f' {attr}="{value}"' for attr, value in sorted(attrs.items()))
    attr_str = ''.join(attr_pairs)
    
    if content:
        elements = (f'<{name}{attr_str}>{c}</{name}>' for c in content)
        return '\n'.join(elements)
    else:
        return f'<{name}{attr_str} />'

print(tag(content='testing', name="img"))  # name goes to name, content goes to attrs scince there is a key word
```

*content (Variable Positional Arguments): The * tells Python to scoop up any extra positional arguments passed to the function and pack them into a tuple called content. If you call tag('p', 'hello', 'world'), 'hello' and 'world' are packed into the content tuple.

**attrs (Variable Keyword Arguments): The ** tells Python to scoop up any extra keyword arguments that don't explicitly match the named parameters in the signature. It packs them into a dictionary called attrs.

```python
my_tag = {'name': 'img', 'title': 'Sunset Boulevard', 'src': 'sunset.jpg', 'class': 'framed'}
tag(**my_tag)
```

## What is Functional Programming?
Functional Programming (FP) is a specific style of writing code.

Normally, when people learn to code, they learn Imperative Programming. You write a step-by-step list of instructions, use for loops, and constantly change the values of your variables to get a result.

Functional Programming flips that on its head. It treats writing software like solving math equations. The entire philosophy boils down to a few strict rules:

Pure Functions: A function should do exactly one thing: take inputs and return an output. It should never change anything outside of itself (no modifying global variables, no writing to a file, no changing the data you passed in). If you put the same input in, you get the exact same output, 100% of the time.

Immutability: Once you create a variable or a list, you are never allowed to change it. If you have a list of numbers and want to double them, you don't change the original list; you create a brand-new list with the doubled numbers.

First-Class Functions: Functions are treated just like regular variables. You can pass a function into another function as an argument, or return a function from a function.

Why do people like it? Because code written this way has zero "side effects." It is incredibly predictable, extremely easy to test, and rarely crashes from weird bugs where a variable got changed when it wasn't supposed to.

Even though Python isn't a strict FP language, Guido and the Python team built functional tools into it so you can use that style when it makes sense.

First-class functions: As mentioned above, Python lets you pass functions around as data. This lets you use functional staples like map(), filter(), and reduce().

Pattern matching: A newer Python feature (the match/case syntax) that lets you rip apart and analyze data structures very cleanly, which is a hallmark of functional languages.

operator and functools: These are built-in Python toolboxes. They provide pre-built functional utilities so you don't have to write them from scratch (which the book is about to teach you).

```python
#The operator Module
from functools import reduce
from operator import mul
def factorial(n):
    return reduce(mul, range(1, n+1))
```

```python
from operator import itemgetter

# A simple list of tuples: (Name, Grade, Age)
students = [
    ('Alice', 'A', 22),
    ('Bob', 'C', 19),
    ('Charlie', 'B', 25)
]

# itemgetter(2) creates a mini-function. 
# It tells Python: "When you look at a student, only pay attention to index 2 (the age)."
sort_by_age = itemgetter(2)

# We pass that mini-function into the 'key' parameter of sorted()
# It will now sort the list based entirely on the ages.
sorted_students = sorted(students, key=sort_by_age)

print(sorted_students)
# Output: [('Bob', 'C', 19), ('Alice', 'A', 22), ('Charlie', 'B', 25)]

# BONUS: You can grab multiple things at once!
# This creates a function that grabs the Grade (index 1) and Name (index 0)
grab_grade_and_name = itemgetter(1, 0)
print(grab_grade_and_name(students[0]))
# Output: ('A', 'Alice')
```

```python
from operator import attrgetter

# 1. Create the "Inner" Class (The Lego brick)
class Owner:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
        
    def __repr__(self):
        # This just makes it print nicely
        return f"Owner({self.name})"

# 2. Create the "Outer" Class (The main structure)
class Dog:
    def __init__(self, dog_name, owner_object):
        self.dog_name = dog_name
        # Here is the Composition! 
        # We are storing an entire 'Owner' object inside the 'Dog' object.
        self.owner = owner_object 
        
    def __repr__(self):
        return f"Dog({self.dog_name}, owned by {self.owner.name})"

# 3. Build our data
# First, we create the Owner objects
owner1 = Owner("Alice", "555-1234")
owner2 = Owner("Charlie", "555-9999")
owner3 = Owner("Bob", "555-5678")

# Next, we create the Dog objects and pass the Owner objects inside them
dogs = [
    Dog("Rex", owner1),      # Rex is owned by Alice
    Dog("Buddy", owner2),    # Buddy is owned by Charlie
    Dog("Max", owner3)       # Max is owned by Bob
]

# 4. Use attrgetter to dig into the nested objects
# We want to sort the dogs alphabetically by their OWNER'S name.
# attrgetter looks at the Dog, finds the '.owner' attribute, 
# and then digs one level deeper to find the '.name' attribute.
sort_by_owner_name = attrgetter('owner.name')

# Sort the list using our digger function
sorted_dogs = sorted(dogs, key=sort_by_owner_name)

# Let's see the result!
for dog in sorted_dogs:
    print(dog)
```

```python
from operator import methodcaller

text = "hello world"

# We create a function that will trigger the '.upper()' method
make_loud = methodcaller('upper')

# When we pass 'text' into it, it runs text.upper()
print(make_loud(text))
# Output: 'HELLO WORLD'

# You can also pre-load arguments into the method.
# This creates a function that will run .replace('world', 'friend')
swap_words = methodcaller('replace', 'world', 'friend')

# We pass 'text' into it, and it executes the replacement
print(swap_words(text))
# Output: 'hello friend'
```

```python
# Freezing Arguments with functools.partial
#In Python, partial takes an existing function, pre-fills (freezes) some of its arguments, and hands you back a brand-new, simpler function that requires less typing to use.

from operator import mul
from functools import partial

# mul() multiplies two numbers: mul(3, 7) -> 21
# If we want to use map() to multiply a whole list of numbers by 3, we have a problem.
# map() only feeds ONE number at a time into the function. It would crash mul().

# Solution: Freeze the first argument of mul() to be the number 3.
triple = partial(mul, 3)

# Now, 'triple' is a brand new function that only needs ONE argument.
print(triple(7))  
# Output: 21 (Because it secretly does mul(3, 7))

# Now it works perfectly with map()
print(list(map(triple, [1, 2, 3, 4])))
# Output: [3, 6, 9, 12]
```

```python
from functools import partial

# tag from previos example

# We freeze the first positional argument to 'img'
# We freeze the keyword-only argument 'class_' to 'pic-frame'
picture = partial(tag, 'img', class_='pic-frame')

# Now, 'picture' is a dedicated function just for making our specific images!
# All we have to provide is the new stuff, like the 'src' link.
print(picture(src='wumpus.jpeg'))
# Output: '<img class="pic-frame" src="wumpus.jpeg" />'
```
