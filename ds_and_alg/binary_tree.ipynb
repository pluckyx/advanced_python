# DFS -> Stack
# BFS -> Queue


```python
class TreeNode:
    def __init__(self,val,left=None,right=None):
        self.val = val
        self.left = left
        self.right= right
    
    def __str__(self):
        return str(self.val)
```

```python
A = TreeNode(1)
B = TreeNode(2)
C = TreeNode(3)
D = TreeNode(4)
E = TreeNode(5)
F = TreeNode(10)

A.left = B
A.right = C
B.left = D
B.right = E
C.left = F

print(A.left)
```

```python
#DFS recursion
def pre_order(node):
    if not node:
        return 
    
    print(node)
    pre_order(node.left)
    pre_order(node.right)

pre_order(A)
```

```python
#DFS recursion
def in_order(node):
    if not node:
        return 
    
    in_order(node.left)
    print(node)
    in_order(node.right)

in_order(A)
```

```python
from collections import deque

def bfs(node):
    q = deque()
    q.append(node)

    while q:
        node = q.popleft()
        print(node)

        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)

bfs(A)
```

```python
def search(node, target):
    if not node:
        return 'no'
    if node.val == target:
        return True

    return search(node.left, target) or search(node.right, target)

search(A,50)
```
