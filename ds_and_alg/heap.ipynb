```python
#Min heap
import heapq

A = [-4,3,1,0,2,5,10,8,12,9]
heapq.heapify(A)
A

```

```python
heapq.heappush(A,4)
```

```python
def heapsort(arr):
    heapq.heapify(arr)
    n = len(arr)
    new_list = [0] * n

    for i in range(n):
        minn = heapq.heappop(arr)
        new_list[i] = minn

    return new_list

heapsort([1,3,5,7,9,2,4,6,8,0])
```

```python
#Max heap (شعبلولي

B = [-4, 0, 1, 3, 2, 5, 10, 8, 12, 9]
n = len(B)

for i in range(n):
    B[i] = -B[i]

heapq.heapify(B)

B
```

```python
-heapq.heappop(B)
```

```python
#Pushing into max
#inserting 7
heapq.heappush(B,-7)
B
```

```python
#slower than heapify n log n
C = [-5, 4, 2, 1 ,7 ,0 ,3]

heap = []

for x in C:
    heapq.heappush(heap,x)
    print(heap)
```

```python
from collections import Counter

D = [2,2,2,2,5,6,4,5,8]

counter = Counter(D)
heap = []
for k,v in counter.items():
    heapq.heappush(heap,(k,v))

heap
```
