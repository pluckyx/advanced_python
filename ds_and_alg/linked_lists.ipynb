# singly linked lists

```python
class SinglyNode():
    def __init__(self,value, next=None):
        self.value = value
        self.next = next
    def __str__(self):
        return str(self.value)
```

```python
Head = SinglyNode(1)
A = SinglyNode(2)
B = SinglyNode(3)
C = SinglyNode(4)

Head.next= A
A.next = B
B.next = C
```

```python
# Travers list
current = Head

while current:
    print(current)
    current = current.next
```

```python
# Display
def display(Head):
    current = Head
    elements = []
    while current:
        elements.append(str(current.value))
        current = current.next
    print (' -> '.join(elements))
display(Head)

```

```python
# Search for value
def search(Head, target):
    current = Head
    while current:
        if current.value == target:
            return True
        current = current.next
    return False

search(Head, 10)
```

# doubly linked lists

```python
class DoublyNode():
    def __init__(self,value, next=None, prev=None):
        self.value = value
        self.next = next
        self.prev = prev
    def __str__(self):
        return str(self.value)
```

```python
head = tail = DoublyNode(1)
```

```python
# Display
def display(head):
    current = head
    elements = []
    while current:
        elements.append(str(current.value))
        current = current.next
    print (' <-> '.join(elements))
display(head)
```

```python
# Insert at beginning
def insert_at_beginning(head,tail,value):
    new_node = DoublyNode(value,next=head)
    head.prev = new_node
    return new_node, tail

head, tail = insert_at_beginning(head, tail, 2)
display(head)
```

```python
# Insert at end
def insert_at_end(head,tail,value):
    new_node = DoublyNode(value,prev=tail)
    tail.next = new_node
    return head, new_node

head, tail = insert_at_end(head, tail, 5)
display(head)
```
