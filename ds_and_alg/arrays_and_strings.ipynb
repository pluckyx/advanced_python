# arrays

```python
A = [1,2,3,4,5]
A
```

```python
#INSERT AT END
A.append(6)
A
```

```python
#DELETING AT END
A.pop()
A
```

```python
#INSERT (NOT AT END) - O(n)

A.insert(2,10)
A
```

```python
#MODIFY
A[3] = 1000
A
```

```python
if 1 in A:
    print(True)

```
