# Hashsets

```python
s = set()
```

```python
# Adding
s.add(1)
s.add(2)
s
```

```python
#Search - O(1)
if 1 in s:
    print(True)
```

```python
#Remove
s.remove(1)
```

```python
s
```

```python
string = 'ssssssssskikkkkkkkkkwwww'
# set don't allow duplicates
s_string = set(string)
s_string
```

```python
#loop over items in set
for i in s:
    print(i)
```

# Hashmaps

```python
d = {'mo' : 1,
     'lo' : 2,
     'ko' : 3}

print(d)
```

```python
#Adding O(1)
d['ro'] = 4
d
```

```python
#Look up key - O(1)

if 'lo' in d:
    print(True)
```

```python
#Look up value
d['mo']
```

```python
for key,value in d.items():
    print(f'key: {key}  val: {value}')
```
