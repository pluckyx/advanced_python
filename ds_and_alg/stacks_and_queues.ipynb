# Queues

```python
# D ouble ended queue u can use it from left or right
from collections import deque

q = deque()
q
```

```python
# add to righ
q.append(2)
q.append(5)
q.append(4)
```

```python
# remove from right
q.pop()
q
```

```python
# remove from right - O(1)
q.popleft()
q
```
