```python
A = [-3,-1,0,1,4,5]

#O(n)
if -1 in A :
    print(True)
```

```python
#O(log n)
def binary_search(arr,target):
    N = len(arr)
    L = 0
    R = N-1

    while L <= R:
        M = L+ (R-L)//2

        if arr[M] == target:
            return True
        elif target < arr[M]:
            R = M-1
        else:
            L = M+1
    return False

binary_search(A,10)



```

```python
# Condition based

B = [False,True,True,False,False,False]
def condition_binary(arr):
    N = len(arr)
    L = 0
    R = N-1
    
    while L < R: #escape equality
        M = L+ (R-L)//2
        if arr[M] == True:
            R = M
        else:
            L = M +1
    return L

condition_binary(B)

        
```
