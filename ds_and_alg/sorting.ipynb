## Bubble sort

```python
A = [-5,3,2,4,5,-8,9]

def bubble_sort(arr):
    n = len(arr)
    flag = True
    while flag:
        flag = False
        for i in range(1,n):
            if arr[i-1] > arr[i]:
                flag = True
                arr[i-1], arr[i] = arr[i], arr[i-1]

    return arr
bubble_sort(A)
```

## Insertion sort


```python
B = [1,2,4,5,-7,-8]

def insertion_sort(arr):
    n = len(arr)
    for i in range(1,n):
        for j in range (i,0,-1):
            if arr[j-1] > arr[j]:
                arr[j-1], arr[j] = arr[j], arr[j-1]
            else:
                break
    return arr
insertion_sort(B)


```

## Merge sort

```python
A = [3,2,1]
def merge(arr):
    if len(arr) <= 1:
        return arr

    m = len(arr) // 2
    L = merge(arr[0:m])
    R = merge(arr[m:])

    i = 0
    j = 0
    k = 0

    while i<len(L) and j<len(R):
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1
    while i <len(L):
        arr[k] = L[i]
        i += 1
        k += 1
    while j <len(R):
        arr[k] = R[j]
        j += 1
        k += 1
    return arr

merge(A)
```

## Quick sort

```python
E = [3,2,1]

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    
    p = arr[-1]
    L = [x for x in arr[:-1] if x <= p]
    R = [x for x in arr[:-1] if x >= p]

    L = quick_sort(L)
    R = quick_sort(R)

    return L + [p] + R

quick_sort(E)
```
