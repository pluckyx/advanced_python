```python
from pyamaze import maze, agent, textLabel

m = maze(10,10)
m.CreateMaze()

for row in range(1, m.rows + 1):
    for col in range(1, m.cols + 1):
        x = m._cell_width * (col - 1) + m._cell_width 
        y = m._cell_width * (row - 1) + m._cell_width 
        
        m._canvas.create_text(
            x, y,
            text=f"({row},{col})",
            font=("Arial", 7)
        )
```

```python
def h(cell1: tuple, cell2=(1,1)) -> int:
    x1, y1 = cell1
    x2, y2 = cell2
    return abs(x1-x2) + (y1-y2)
```

```python
from queue import PriorityQueue

def astar(m: maze)->dict:
    start = (m.rows, m.cols)
    g_score = {cell: float('inf') for cell in m.grid}
    g_score[start] = 0
    f_score = {cell: float('inf') for cell in m.grid}
    f_score[start] = h(start, (1,1))

    came_from = {}  #we are going backwards to ignore the explored childs not used
    open_set = PriorityQueue()
    open_set.put((h(start,(1,1))+g_score[start],h(start,(1,1)),start))

    while not open_set.empty():
        current = open_set.get()[2]
        if current == (1,1):
            break

        for d in 'ESNW':
            if m.maze_map[current][d] == True:
                if d == 'E':
                    child = (current[0], current[1]+1)
                if d == 'W':
                    child = (current[0], current[1]-1)
                if d =='N':
                    child = (current[0]-1, current[1])
                if d == 'S':
                    child = (current[0]+1, current[1])
                
                temp_g_score = g_score[current] +1
                temp_f_score = temp_g_score + h(child)

                if temp_f_score < f_score[child]:   #prob infinite rn
                    g_score[child] = temp_g_score
                    f_score[child] = temp_f_score
                    open_set.put((temp_f_score,h(child),child))
                    came_from[child] = current

    forward_path = {}
    cell = (1,1)

    while cell!=start:
        forward_path[came_from[cell]] = cell
        cell = came_from[cell]
    
    return forward_path

path = astar(m)
a= agent(m, color='red', filled=True)
m.tracePath({a:astar(m)})
l = textLabel(m, 'A Star Path Length', len(path)+1)
m.run()

                

            

    
```
